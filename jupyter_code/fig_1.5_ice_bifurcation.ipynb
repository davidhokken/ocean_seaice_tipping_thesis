Now we draw the bifurcation diagram of $I$:

```python
# First, we find the equilibrium values of $I$ by using the pseudo-arc continuation:
I_init = -2.8
R_init = -0.35
s_end = 5.5
ds = 0.01
I = psarc(I_init, R_init, s_end, ds)

# Next, we find the indices i1 and i2 where dI/dR changes sign:
for i in range(len(I)-2):
    if I[i+2][0] < I[i+1][0]:
        if I[i+1][0] > I[i][0]:
            i1 = i
            
for i in range(len(I)-2):
    if I[i+2][0] > I[i+1][0]:
        if I[i+1][0] < I[i][0]:
            i2 = i

# Now we are ready to plot the figure
plt.figure(figsize=(7,5))
plt.plot([Ival[0] for Ival in I[:i1]], [Ival[1] for Ival in I[:i1]], color='b', linewidth=2)
plt.plot([Ival[0] for Ival in I[i1:i2]], [Ival[1] for Ival in I[i1:i2]], color='k', ls=":", linewidth=2)
plt.plot([Ival[0] for Ival in I[i2:]], [Ival[1] for Ival in I[i2:]], color='r', linewidth=2)
plt.xlabel('$R$')
plt.ylabel('$I$')
plt.text(0.1, 1.05, 'ice-covered', color = 'r', weight = 'bold')
plt.text(-0.25, -1.75, 'ice-free', color = 'b', weight = 'bold')
plt.show()
```
